```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

### Taking the total spend 

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

### Checking the relationship between total spend and sales

```python
sns.scatterplot(x = 'total_spend',y = 'sales',data = df)
```

### Separating the x and y parameters

```python
X = df['total_spend']
y = df['sales']
```

### This is the regression line.

```python
sns.regplot(x = 'total_spend',y = 'sales',data = df)
```

### It gives the beta coefficient and it returns the highest order coefficient first

```python
np.polyfit(X,y,deg=1)
```

```python
# Future spend budgets
potential_spend = np.linspace(0,500,100)
```

### Predicted Sales according the potential spend

```python
predicted_sales = 0.04868788*potential_spend + 4.24302822
```

```python
plt.plot(potential_spend,predicted_sales)
```

### We are getting a similar Plot as the  regplot

```python

sns.scatterplot(x = 'total_spend',y = 'sales',data = df)
plt.plot(potential_spend,predicted_sales,color = 'red')
```

### Now i will put any value for potential spend and check the predicted sales for it.

```python
spend = 200
predicted_sales = 0.04868788*spend + 4.24302822
```

```python
predicted_sales
```
